```python
# installing dependencies 

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
import sklearn
import mlflow
```

```python
# load the dataset
df = pd.read_csv("stock_market_dataset.csv")
df.info()
```

Perform data anlysis 

```python
summary_stats = df.describe()
correlation_matrix = df.corr()
plt.figure(figsize = (10,8))
sns.heatmap(correlation_matrix, annot = True, cmap ='coolwarm')
plt.title('Correlation Matrix')
plt.savefig('Correlation_matrix.png')
```

```python
# Log analysis in MlFlow
mlflow.set_experiment("Stock Analysis")
with mlflow.start_run():
    mlflow.log_param("data file", df)
    mlflow.log_artifact('Correlation_matrix.png')
    mlflow.log_text(summary_stats.to_string(), "summary_stats.txt")
    mlflow.log_text(correlation_matrix.to_string(),"correlation_matrix.txt")

print("Data analysis logged in MLflow experiment")
```
